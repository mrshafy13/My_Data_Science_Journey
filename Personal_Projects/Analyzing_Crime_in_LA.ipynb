```python
# Re-run this cell
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv("crimes.csv", parse_dates=["Date Rptd", "DATE OCC"], dtype={"TIME OCC": str})
crimes.head()
```

```python
import datetime
from datetime import datetime 
hours = []
for time in crimes['TIME OCC']:
    datetime_obj = datetime.strptime(time, "%H%M")
    hour = datetime_obj.time().hour
    hours.append(hour)
```

```python
crimes.insert(2, "Time_Hour", hours)
```

```python
peak_crime_hour = 0
count = 0
for hour in range (max(hours)+1):
    a = hours.count(hour)
    if(count<a):
        count = a
        peak_crime_hour = hour
    hour = hour+1
peak_crime_hour
```

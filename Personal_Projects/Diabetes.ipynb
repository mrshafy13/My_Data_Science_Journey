### Finding corelation and visualizing using heatmap

```python
diabetes.corr()
```

```python
sns.heatmap(diabetes.corr())
plt.show()
```

### Counting the value of the column 'Outcome' and visualizing using Barplot

```python
diabetes[['Outcome']].value_counts()
```

```python
sns.barplot(x= 'Outcome', y = diabetes['Outcome'].value_counts(), data = diabetes)
plt.show()
```

### Visualizing 'Age' column using boxplot

```python
sns.boxplot(diabetes['Age'])
plt.show()
```

### Minimum, Maximum and mean value of 'BMI' column

```python
print(diabetes[['BMI']].min())
print(diabetes[['BMI']].max())
print(diabetes[['BMI']].mean())
```

### Scatter Plot of 'Glucose' vs 'BMI'

```python
sns.scatterplot(x='Glucose', y='BMI', data = diabetes, color = 'green')
plt.title("'Glucose' vs 'BMI'")
plt.xlabel("Glucose")
plt.ylabel("BMI")
plt.show()
```

### Histogram of All Attribute

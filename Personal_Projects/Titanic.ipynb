### Importing Dataset to Dataframe

```python
URL = 'https://drive.usercontent.google.com/u/0/uc?id=10WTiQLR6-e-KlxcYRhSFP_gEBVsHCMKN&export=download'
```

```python
titanic = pd.read_csv(URL)
print("Size =", titanic.size)
print("Shape =", titanic.shape)
```

### Retrieving 20 random data from dataset

```python
titanic.sample(n=20)
```

### Printing the total number of null value in the dataset

```python
titanic.isnull().sum().sum()
```

We can see there's 866 null values, we need to fill them.

```python
titanic.describe(include='all')
```

From the statistic description, most of the values are missing from Cabin column and some of them are from Age column

### Replacing Null values using forward fill method

```python
titanic['Age'].fillna(method = 'ffill', inplace = True)
titanic['Cabin'].fillna(method = 'ffill', inplace = True)
titanic['Embarked'].fillna(method = 'ffill', inplace = True)
```

```python
titanic.describe(include='all')
```

All the missing values are replaced using forward fill method
